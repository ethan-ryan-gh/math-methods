# Higher-Order Ordinary Differential Equations

Python has many excellent functions for solving Ordinary Differential Equations. One example, and the focus of our notebook, is `odeint`. To use `odeint` for higher order differential equations we need to do three things:

1) Rewrite our problem as a system of first-order ODEs
2) Define a model which we will pass to `odeint`
3) Provide `odeint` with our initial conditions


Let's look at an example using the ODE y'' + 2y' + y = 0.

First thing we need to do is rewrite as a system of first-order ODEs, which means we need one or more differential equations where we only take the first derivative. We can do this by introducing the new variables y1 and y2:

y1 = y
y2 = y'


Looking back at our original ODE, we can define y1 and y2 as:

y1' = y2
y2' = -2y2 - y1


The next thing we need to do is define our model. This is done in the example code below. Here we are rewriting our differential equation in terms of y1 and y2 in a way that `odeint` can process:


```python

def model(y, t):
    y1, y2 = y
    dy1dt = y2
    dy2dt = -2*y2 - y1
    return [dy1dt, dy2dt]

```

Now, let's bring it all together and use `odeint` to solve our differential equation:

```python
from scipy.integrate import odeint
import numpy as np

# Initial conditions
y0 = [1, 0]  # y(0) = 1, y'(0) = 0

# Time points
t = np.linspace(0, 10, 100)

# Solve the ODE
sol = odeint(model, y0, t)

# Access the solution
y = sol[:, 0]  # y(t)
y_prime = sol[:, 1]  # y'(t)
```

In our model, y1 or y is our first arguement, and y2 or y' is the second, so we pass our initial conditions in that order when defining y0. `Sol` is the name of the array that stores the solutions to our system of first-order differential equations (we have one for y, one for y,). The y solutions comes first (it is the first one in the model), and can be found in the zeroth column of `sol` (this is what the command `y=sol[:,0]` does - extracts y(t) from the first column, using the values from all the rows. Python array indexing begins at 0.

We can now use `matplotlib` to visulaize our solution:

```python
import matplotlib.pyplot as plt

plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution to the ODE')
plt.grid()
plt.show()
```

Let's apply the above procedure to a new example - the Damped Oscillator. 

Modify the code below to include the initial conditions y(0) = 1, y'(0) = 0, and make a plot of the solution y(t) (you can copy the code above).

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def damped_oscillator(y, t, m, c, k):
    y1, y2 = y
    dy1dt = y2
    dy2dt = -k/m * y1 - c/m * y2
    return [dy1dt, dy2dt]

m = 1  # Mass
c = 0.5  # Damping coefficient
k = 1  # Spring constant

y0 = [1, 0]
t = np.linspace(0, 20, 100)

sol = odeint(damped_oscillator, y0, t, args=(m, c, k))

plt.plot(t, sol)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution to the ODE')
plt.grid()
plt.show()
```
