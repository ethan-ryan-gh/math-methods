# Fourier Series 2

The fourier series for a square wave is $$ f(x) = \frac{4}{\pi} \sum \limits^{\infty}_{n = odd} \frac{1}{n}{\rm sin}\big( \frac{n \pi x}{L} \big) $$


Let us explore how this summation converges. The function below will plot the Fourier Series above for a given number of terms or values of n(`N`). L is the period of oscillation or width of the repeating unit. `num_points` is used in creating the mesh to generate the plot (does not need to be changed). Generate 3 separate graphs showing the Fourier Series calculated with N= 5, N=20, and N=100. Include all three graphs in this notebook. 

Below is an example with N=1 (ie we take only the first term in the Fourier Series).


```python

import matplotlib.pyplot as plt
from numpy import pi,sin,linspace,zeros,arange,cos


def square_wave(N = 1,L = 5,num_points = 200):
    x = linspace(-4*pi,4*pi,num_points)
    y = zeros(num_points)
    for n in arange(1,N+1,2):
        y = y + (1.0/n * sin(n*pi*x/L))
    y = y * 4.0/pi
    plt.plot(x,y)
```

```python

square_wave(N=1,L=5)
```

```python
square_wave(N=5,L=5)
```

```python
square_wave(N=20,L=5)
```

```python
square_wave(N=100,L=5)
```

```python
square_wave(N=5,L=2)
```

```python
square_wave(N=5,L=10)
```

Provide detailed comments on how the Fourier Series changes as the number of points N is increased. What happens when you change L?

Well in the most basic form, when we increase n it increases the number of squibbles. Mathematically, it increases the number of peaks, or oscillations over a given distance L. At n=1 there's only 1 peak for every oscillation, but for n=5 there's 3 peaks for every oscillation. What this is doing is making the fourier series more accurate to the step function we're trying to solve. By n=100 it barely even looks like a sum of sines, it just about looks purely square.

When we change L, as I did above, it changes the period T, or the number of oscillations within a given length. As we saw when I changed L to 10, the number of steps in the square wave decreased. Vice versa for when I changed L to 2 the number of oscillations increased.
